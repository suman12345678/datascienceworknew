```python
import pandas as pd
dataset=pd.read_csv("cancersurvival.csv",header=None)
names = ['Age', 'Year operation', 'Axillary nodes detected', 'Survival status']
dataset.columns=names
#dataset.head(5)
print('total number of row and columns'+ str(dataset.shape))
```

# Attribute Information:
Age of patient at time of operation (numerical)

Patient's year of operation (year - 1900, numerical)

Number of positive axillary nodes detected (numerical)

Survival status (class attribute)
  
  1 = the patient survived 5 years or longer
  
  2 = the patient died within 5 year



```python
a=dataset['Survival status'].value_counts()
print('patient dieded within 5 years'+str(a[2]))
print('patient survived 5 years or longer'+str(a[1]))
```

# Objective
1. To find out patter of cancer survival with Age of patience, Patient's year of operation, Number of +ve auxilary node detected

```python
import seaborn as sns
import matplotlib.pyplot as plt
#names = ['Age', 'Year operation', 'Axillary nodes detected', 'Survival status']
sns.set_style("whitegrid")
sns.FacetGrid(dataset,hue='Survival status',size=4) \
 .map(plt.scatter,"Age", "Year operation").add_legend()
plt.show()    

```

```python
sns.set_style("whitegrid")
sns.FacetGrid(dataset,hue='Survival status',size=4) \
 .map(plt.scatter,"Year operation", "Axillary nodes detected").add_legend()
plt.show()    


```

```python
#pairplot difficult for many feature(use pca tsne for many catagory)
sns.set_style("whitegrid")
sns.pairplot(dataset,hue="Survival status",size=4) 
plt.show()
#names = ['Age', 'Year operation', 'Axillary nodes detected', 'Survival status']
```

# finding
with age more people survived less
with more year of operation people survived more
with more auxilary node detected people survived less
