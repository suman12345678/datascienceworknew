#### single Linked list implementation

```python
class Node(object):  # a node object has data and next node object
    
    def __init__(self,data):
        self.data=data
        self.nextNode=None
        
class LinkedList():  # a linkedlist object has attribute head node and size
    
    def __init__(self):
        self.head=None
        self.size=0
        
    #O(1)
    def insertStart(self,data):
        self.size=self.size+1  # increase size   
        newNode=Node(data)     # create a node new object with nextNode as None
        if not self.head:  # if first time
            self.head=newNode
        else: # if already node/nodes exists
            newNode.nextNode=self.head  # set current head as nextnode
            self.head=newNode  # set current node as head
    
    # O(1)
    def size1(self):
        return self.size
    
    #O(N) not good
    def size2(self): # alternate find size 
        actualNode=self.head
        size=0
        while actualNode is not None: # loop from head to last node
            size+=1
            actualNode=actualNode.nextNode
        return size    
    
    #O(N) insert end
    def insertEnd(self,data):
        self.size=self.size+1  # increase size
        newNode=Node(data)  # create new node object with nextNode as None
        currentNode=self.head  # starfrom head
        if currentNode is not None:
            while currentNode.nextNode is not None:
                currentNode=currentNode.nextNode
            currentNode.nextNode=newNode
        else:
            self.head=newNode
    
    def traverseList(self):
        currentNode=self.head
        while currentNode is not None:
            print(currentNode.data)
            currentNode=currentNode.nextNode
    
    def remove(self,data):
        if self.head is None:
            return
        self.size=self.size-1
        currentNode=self.head
        previousNode=None
        while currentNode.data !=data:
            previousNode=currentNode
            currentNode=currentNode.nextNode
        if previousNode is None:
            self.head=currentNode.nextNode
        else:
            previousNode.nextNode=currentNode.nextNode
            
L=LinkedList()
print('add 10 at end')
L.insertEnd(10)
print('add 9 at start')
L.insertStart(9)
L.traverseList()
print('add 10 at end')
L.insertEnd(10)
L.traverseList()
print('size1',L.size1())
print('size2',L.size2())
print('remove 9')
L.remove(9)
L.traverseList()
            
```

#### Implement stack

```python
class Stack:    
    def __init__(self):
        self.stack=[]
        
    def isEmpty(self):
        return self.stack==[]
    
    def pop(self):
        data=self.stack[-1]   #self.stack[0] for queue
        del self.stack[-1]
        return data
    
    def peak(self):
        data=self.stack[-1]
        return data
    
    def push(self,data):
        self.stack.append(data)
        
    def sizeStack(self):
        return len(self.stack)
    
    def showstack(self):
        for i in self.stack:
            print(i)
    
s=Stack()
print('push 1')
s.push(1)
print('push 2')
s.push(2)
print('push 3')
s.push(3)
print('size',s.sizeStack())
print('Stack items are ',s.showstack)
print('pop last item ')
s.pop()
print('size',s.sizeStack())
print('Stack items are ',s.showstack)

```

#### Implement binary tree

```python
class Node(object):
    
    def __init__(self,data):# define node with leftChild and rightChild as None    
        self.data=data
        self.leftChild=None
        self.rightChild=None
        
class BinarySearchTree(object):
     
    def __init__(self): # initialize a tree with root=None
        self.root=None
            
    def insert(self,data):
        if not self.root:  # if root doesnot exist create while first time inserting data into tree
            self.root=Node(data)
        else:
            self.insertNode(data,self.root)
            
    def insertNode(self,data,node):
        if data<node.data: # If data is left side of a node
            if node.leftChild: # if the nodehas leftchild call recursively
                self.insertNode(data,node.leftChild)
            else:  # if there is no leftchild of the current node then create a child on left
                node.leftChild=Node(data)
        
        if data>node.data: # If data is right side of a node
            if node.rightChild: # if the nodehas rightchild call recursively
                self.insertNode(data,node.rightChild)
            else:  # if there is no rightchild of the current node then create a child on right
                node.rightChild=Node(data)
        
    def getMinValue(self):
        if self.root:
            return self.getMin(self.root)
        
    def getMin(self,node):
        if node.leftChild:
            return self.getMin(node.leftChild)
            
        return node.data    
        
    def getMaxValue(self):
        if self.root:
            return self.getMax(self.root)
        
    def getMax(self,node):
        if node.rightChild:
            return self.getMax(node.rightChild)
            
        return node.data    
    
    def traverse(self):
        if self.root:
            self.traverseInOrder(self.root)
            
    def traverseInOrder(self,node):
        if node.leftChild:
            self.traverseInOrder(node.leftChild)
        print(node.data)
        if node.rightChild:
            self.traverseInOrder(node.rightChild)
            
    def remove(self,data):
        if self.root:
            self.root=self.removeNode(data,self.root)  # pass data to be deleted and root node 
            
    def removeNode(self,data,node):
        if not node:
            return node
        
        if data<node.data:
            node.leftChild=self.removeNode(data,node.leftChild)  #replace the leftChild of node with new value returned from removeNode
        elif data>node.data:
            node.rightChild=self.removeNode(data,node.rightChild) #replace the rightChild of node with new value returned from removeNode
        else:  # Here located the data in binary tree and base condition of recursive function
            if (not node.leftChild) and (not node.rightChild):  # Its a leaf node
                print('Removing leaf node')
                del node
                return None
            if not node.leftChild: #remove a node with single right child
                print('Removing a node with single rightchild')
                tempNode=node.rightChild  
                del node
                return tempNode  #return 
            elif not node.rightChild: #remove a node with single left child
                print('Removing a node with single leftchild')
                tempNode=node.leftChild  
                del node
                return tempNode  #return 
            
            #remove node with 2 children
            print('removing node with 2 children')
            tempNode=self.getPredeccor(node.leftChild)  # get max value on whole left side of located node
            node.data=tempNode.data  # just replace the node.data with tempNode.data
            node.leftChild=self.removeNode(tempNode.data,node.leftChild)  # now delete the tempNode in left subtree recursively 
        
        return node    
            
    def getPredeccor(self,node):
        if node.rightChild:
            return self.getPredeccor(node.rightChild)
        return node
            
            
        
b=BinarySearchTree()
print('insert 1')
b.insert(1)
print('insert 2')
b.insert(2)
print('insert 20')
b.insert(20)
print('insert 30')
b.insert(30)
print('insert 4')
b.insert(4)
print('insert 40')
b.insert(40)
print('insert 400')
b.insert(400)
print('insert 41')
b.insert(41)
print('Min value',b.getMinValue())
print('Max value',b.getMaxValue())
print('items in sorted order',b.traverse())
        
            
        
```
